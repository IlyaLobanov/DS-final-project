Порисуем сначала свечи. 

```python
btc_plot = btc[['timestamp', 'open', 'high', 'low', 'close']].copy()

btc_plot['timestamp'] = pd.to_datetime(btc_plot['timestamp'])
btc_plot.set_index('timestamp', inplace=True)


# Создаем график
mpf.plot(btc_plot, type='candle', style='yahoo', title='Minute Candlestick Chart BTC', ylabel='Price', figsize=(20, 10))
```

Прогоним данные через наш бектестер

```python
%autoreload 2

from backtest import Backtester
# Пример использования
symbol = 'BTCUSDT'
candles_data = btc[['timestamp', 'open', 'high', 'low', 'close']] 
initial_cash = 100000
commission = 0.001
backtester = Backtester(symbol, candles_data, initial_cash, commission)


```

```python
backtester.run_backtest('SMA', 2)
```
